# Load Volume and Display image

change to parent directory to use aitom library

load mrc file using io module

example data: 

http://ftp.ebi.ac.uk/pub/databases/empiar/archive/10227/data/US1363_G1/14nov13a_tilt/14nov13a__001.mrc.bz2

http://ftp.ebi.ac.uk/pub/databases/empiar/archive/10227/data/US1363_G1/14nov13a_tilt/14nov13a__002.mrc.bz2


```python
import aitom.io.file as io_file
a = io_file.read_mrc_data("data/14nov13a__001.mrc.bz2")
```

display image using image module

```python
import aitom.image.vol.util as im_vol_util
a_im = im_vol_util.cub_img(a)
```

**a_im** is a dict:
-    'im': image data
-    'vt': volume data

image data is type of numpy.ndarray, elements $\in$ [0, 1]

```python
print(type(a_im['im']))
print(a_im['im'].shape)
print(a_im['im'][1][1])
```
